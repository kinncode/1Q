Pandas中的資料型別
可將資料分為連續(continuous)資料與類別(categorical)資料。
• 連續資料的型別為數值,代表某測量值,例如身高、體重、時速,連續資料的值可以有無限多種可能性。
• 類別資料是離散的(可能值有限),例如血型、性別、居住地。
Pandas中常見的資料型別:
• float: 浮點數型別
• int: 整數型別
• object: 用來儲存字串或混合型別(例如:同時存有數值和字串資料的欄位)的資料
• category: 用來表示類別資料
• bool/boolean:布林變數,用於邏輯判斷
√ 資料匯入DataFrame後,要確認每一欄位的資料型別,因資料型別會影響處理資料的方式。
• Pandas數值資料(整數和浮點數)的預設精確度為64位元。
• 可以用dtypes檢視每一欄位的資料型別

Detecting Outliers
Z-Score
z = (x-mean) / std
To identify outliers using the z-score, we can set a threshold value, say 3. Any data point with a z-score greater than 3 or less than -3 can be
considered an outlier.

```python
np.random.seed(55)
datal= np.concatenate (([-140, -125],
    np.random. choice (np. arange (150, 180), size=50, replace=True),
    [400, 450]))
```

Interquartile Range (IQR)
The interquartile range (IQR) is a measure of the spread of the middle 50% of the data. The IQR can be calculated as the differen the 75th percentile and the 25th percentile of the dataset. Any data point outside the range of 1.5 times the IQR below the 25th p above the 75th percentile can be considered an outlier.


```python
 # calculate IQR for column Height
Q1 = df ['Height']. quantile (0.25)
Q3 = df ['Height']. quantile (0.75)
IQR = Q3- Q1
```

```python
IQR
```

```python
threshold = 1.5
outliers = df[(df ['Height'] < Q1-  threshold* IQR) | (df ['Height'] > Q3 + threshold * IQR)]
```

```python
outliers
```

Excluding Outliers

```python
df1 = df
```

```python
df = df.drop (outliers. index)
df
```
